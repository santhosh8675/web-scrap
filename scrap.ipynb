```python
def rating_of_drug(name):
    name=name.lower()
    url='http://www.druglib.com/ratingsreviews/'+name+'/'
    response=requests.get(url)
    soup=bs4.BeautifulSoup(response.content,'html.parser')
    rating_data=soup.findAll('div',attrs={'class':'info_box'})
    for i in rating_data:
        string=i.b.text
    return string.split()[3]
```

```python
rating_of_drug(name)
```
